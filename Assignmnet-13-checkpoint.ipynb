### Q.1. **Create two int type variables, apply addition, subtraction, division, and multiplication, and store the results in variables. Then print the data in the following format by calling the variables:**

```python
# Create two integer variables
first_var = 10
second_var = 5

# Perform the arithmetic operations
addition = first_var + second_var
subtraction = first_var - second_var
multiplication = first_var * second_var
division = first_var / second_var

# Print the results
print(f"First variable is {first_var} & second variable is {second_var}.")
print(f"Addition: {first_var} + {second_var} = {addition}")
print(f"Subtraction: {first_var} - {second_var} = {subtraction}")
print(f"Multiplication: {first_var} * {second_var} = {multiplication}")
print(f"Division: {first_var} / {second_var} = {division}")
```

**Output:**
```
First variable is 10 & second variable is 5.
Addition: 10 + 5 = 15
Subtraction: 10 - 5 = 5
Multiplication: 10 * 5 = 50
Division: 10 / 5 = 2.0
```

### Q.2. **What is the difference between the following operators:**

(i) **‘/’ & ‘//’**
- **‘/’ (Division Operator)**: This operator performs floating-point division, which always returns a floating-point number.
  - **Example**: `7 / 2` results in `3.5`
  
- **‘//’ (Floor Division Operator)**: This operator performs integer division and returns the largest integer less than or equal to the result (floor of the division).
  - **Example**: `7 // 2` results in `3`

(ii) **‘**’ & ‘^’**
- **‘**’ (Exponentiation Operator)**: This operator raises the left operand to the power of the right operand.
  - **Example**: `2 ** 3` results in `8`
  
- **‘^’ (Bitwise XOR Operator)**: This operator performs a bitwise exclusive OR (XOR) operation on two integers. It compares corresponding bits of the operands, returning `1` for a bit if the operands are different and `0` if they are the same.
  - **Example**: `5 ^ 3` results in `6` (binary: `101 ^ 011 = 110`)

### Q.4. **Explain right shift operator and left shift operator with examples.**

- **Left Shift Operator (`<<`)**: Shifts the bits of a number to the left by a specified number of positions. Each left shift operation effectively multiplies the number by 2.
  
  **Example**:
  ```python
  x = 5  # Binary: 101
  result = x << 1  # Shift left by 1 position -> 1010 (Binary) = 10 (Decimal)
  print(result)  # Output: 10
  ```

- **Right Shift Operator (`>>`)**: Shifts the bits of a number to the right by a specified number of positions. Each right shift operation effectively divides the number by 2.
  
  **Example**:
  ```python
  x = 10  # Binary: 1010
  result = x >> 1  # Shift right by 1 position -> 101 (Binary) = 5 (Decimal)
  print(result)  # Output: 5
  ```
